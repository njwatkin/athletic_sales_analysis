```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_2020 = pd.read_csv('athletic_sales_2020.csv')
athletic_2021 = pd.read_csv('athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_2020.head()
```

```python
# Display the 2021 sales DataFrame
athletic_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_2020.dtypes
```

```python
# Check the 2021 sales data types.
athletic_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_2020_2021_df = pd.concat([athletic_2020, athletic_2021], keys = ["Year 2020", "Year 2021"],axis=0)
athletic_2020_2021_df.head()
athletic_2020_2021_df.reset_index()
```

```python
# Check if any values are null with count()
athletic_2020_2021_df.count()
```

```python
# Check if any values are null using isnull().sum())
null_values = athletic_2020_2021_df.isnull().sum()
null_values
```

```python
# Check the data type of each column
column_type = athletic_2020_2021_df.dtypes
column_type
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_2020_2021_df['invoice_date'] = pd.to_datetime(athletic_2020_2021_df['invoice_date'], format='%Y-%m-%d')
athletic_2020_2021_df.head(10)
athletic_2020_2021_df.dtypes
```

```python
# Confirm that the "invoice_date" data type has been changed.
#See above all in one step.
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
athletic_sales_sum = athletic_2020_2021_df.groupby(['region', 'state', 'city']).agg({'units_sold': 'sum'})

# Rename the sum to "Total_Products_Sold".
athletic_sales_sum.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)


# Show the top 5 results.
df_sortedgb = athletic_sales_sum.sort_values('Total_Products_Sold', ascending = False)
df_sortedgb.head(5)
                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
athletic_sales_sum_pivot = pd.pivot_table(athletic_2020_2021_df, values = 'units_sold', index = ['region', 'state', 'city'], aggfunc = 'sum')

# Rename the "units_sold" column to "Total_Products_Sold"
athletic_sales_sum_pivot.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
df_sortedpiv = athletic_sales_sum_pivot.sort_values('Total_Products_Sold', ascending = False)
df_sortedpiv.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
athletic_sales_sum = athletic_2020_2021_df.groupby(['region', 'state', 'city']).agg({'total_sales': 'sum'})

# Rename the "total_sales" column to "Total Sales"
athletic_sales_sum.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
df_sortedgb = athletic_sales_sum.sort_values('Total Sales', ascending = False)
df_sortedgb.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
athletic_sales_sum_pivot = pd.pivot_table(athletic_2020_2021_df, values = 'total_sales', index = ['region', 'state', 'city'], aggfunc = 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
athletic_sales_sum_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
df_sortedpiv = athletic_sales_sum_pivot.sort_values('Total Sales', ascending = False)
df_sortedpiv.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
df_womens_afw = athletic_2020_2021_df[athletic_2020_2021_df['product'].str.contains("Women's Apparel")].reset_index()
df_womens_afw
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
athletic_sales_sum_wafw_gb = df_womens_afw.groupby(['retailer','region', 'state', 'city']).agg({'units_sold': 'sum'})

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
athletic_sales_sum_wafw_gb.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)


# Show the top 5 results.
df_sortedgb = athletic_sales_sum_wafw_gb.sort_values('Womens_Footwear_Units_Sold', ascending = False)
df_sortedgb.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
athletic_sales_sum_wafw_piv = pd.pivot_table(df_womens_afw, values = 'units_sold', index = ['retailer','region', 'state', 'city'], aggfunc = 'sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
athletic_sales_sum_wafw_piv.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
df_sortedpv = athletic_sales_sum_wafw_piv.sort_values('Womens_Footwear_Units_Sold', ascending = False)
df_sortedpv.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_afw_day_piv = pd.pivot_table(df_womens_afw, values = 'total_sales', index = 'invoice_date')

# Optional: Rename the "total_sales" column to "Total Sales"
womens_afw_day_piv.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
df_sortedpiv = womens_afw_day_piv.sort_values('Total Sales', ascending = False)
df_sortedpiv.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales_wafw = womens_afw_day_piv.resample('D').sum()
daily_sales_wafw

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales_wafw_sorted = daily_sales_wafw.sort_values(by="Total Sales", ascending=True).round(2)
daily_sales_wafw_sorted
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weeklybin_sales_wafw = womens_afw_day_piv.resample('W').sum()
weeklybin_sales_wafw

# Sort the resampled pivot table in ascending order on "Total Sales".
weeklybin_sales_wafw_sorted = weeklybin_sales_wafw.sort_values(by="Total Sales", ascending=True).round(2)
weeklybin_sales_wafw_sorted
```
